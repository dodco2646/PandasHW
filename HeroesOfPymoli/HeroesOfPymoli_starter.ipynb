```python
#age_df = purch_df['SN'].unique()
gender_df = purch_df[["SN","Gender","Age"]]
gender_df_rm_dups = gender_df.drop_duplicates(subset=['SN'])
age_df = gender_df_rm_dups[['Age']].copy()

bins = [0,9,14,19,24,29,34,39,100]
groups = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

age_df["Age_Group"] = pd.cut(age_df.Age,bins,labels=groups)

age_demographics = age_df.groupby('Age_Group')
age_demographics1 = age_demographics.count()
age_demographics1['Percentage of Players'] = age_demographics1['Age'] /len(purch_df['SN'].unique())*100
age_demographics1
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
MPI_df = purch_df[['Item ID','Item Name', 'Price']]
MPI_gb = MPI_df.groupby(['Item ID','Item Name'])

MPI_Final = pd.DataFrame({'Purchase Count':MPI_gb['Price'].count(),
                          'Item Price': MPI_gb['Price'].max(),
                          'Total Purchase Value': MPI_gb['Price'].sum()
                         
})

MPI_Final.sort_values(['Purchase Count'],ascending=False).head()
```
